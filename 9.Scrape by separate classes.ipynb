```python
class Twitter():
    def __init__(self):
        pass
    def scrape_twitter(self,keyword,number_of_tweets):
        self.keyword = keyword
        self.number_of_tweets = number_of_tweets
        scraping_the_tweets = sntwitter.TwitterSearchScraper(str(self.keyword)).get_items()
        
        #Now slice the tweets, let number_of_tweets=100 to slice the first 100 tweets, increase the number to slice more tweets
        slicing_the_tweets = itertools.islice(scraping_the_tweets,int(self.number_of_tweets))
        
        #Creating the dataframe of sliced tweets
        df = pd.DataFrame(slicing_the_tweets)[['date','content']]
        return df
```
